```python
from random import choices
from typing import List

from pandas import read_csv, DataFrame
from seaborn import heatmap
```

```python
dataframe = read_csv('../data/raw/marketing_campaign.csv')
dataframe.drop(columns=['Z_Revenue', 'Z_CostContact'], inplace=True)
NUMERICAL_COLUMNS = [column for column in dataframe.columns if dataframe[column].dtype in ('int64', 'float64')]
```

```python
def custom_heatmap(dataframe: DataFrame, columns: List[str], annot: bool = False) -> None:
    """
    Create correlation heatmap of specifics numericals columns from a dataframe
    Input:
        dataframe, Dataframe: The dataset to chart
        columns, List[str]: The list of features selected from the dataset
        annot, bool = False: If true, the correlation indice will be shown
    """
    dataframe_matrix = dataframe[columns].corr()
    heatmap(dataframe_matrix, annot=annot)
```

```python
random_features = choices(NUMERICAL_COLUMNS, k=5)
custom_heatmap(dataframe, random_features, annot=True)
```
